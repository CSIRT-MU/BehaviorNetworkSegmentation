## Compare implementations

### Import dataset

```python
df = pd.read_pickle('../dataset/balanced_preprocessed_dataset.pkl')
df.head()
```

```python
%%cython -a

import numpy as cnp
cimport numpy as cnp
from libc.math cimport sqrt, pow
cnp.import_array()

boundscheck=False
wraparound=False
cdef double minimum(double[:]  s, int i, int size, int r):

    cdef int lower = max(0, i - r)
    cdef int upper = min(size, i + r + 1)
    cdef int j
    cdef double m = s[lower]
    for j in range(lower, upper):
        if s[j] < m:
            m = s[j]
    return m

boundscheck=False
wraparound=False
cdef double maximum(double[:]  s, int i, int size, int r):

    cdef int lower = max(0, i - r)
    cdef int upper = min(size, i + r + 1)
    cdef int j
    cdef double m = s[lower]
    for j in range(lower, upper):
        if s[j] > m:
            m = s[j]
    return m  

boundscheck=False
wraparound=False
def LB_Keogh(double[:] s1, double[:] s2, int size, int r):
    cdef double LB_sum = 0
    cdef double curr, lower, upper
    cdef int i
    for i in range(size):
        
        curr = s1[i]
        lower = minimum(s2, i, size, r)
        upper = maximum(s2, i, size, r)

        if curr > upper:
            LB_sum += pow((curr-upper), 2.0)
        elif curr < lower:
            LB_sum += pow((curr-lower), 2.0)

    return sqrt(LB_sum)
```

```python
def LB_Keogh_py(s1,s2,r):
    
    LB_sum=0
    
    for ind,i in enumerate(s1):
        
        lower_bound= min(s2[(ind-r if ind-r>=0 else 0):(ind+r+1)])
        upper_bound= max(s2[(ind-r if ind-r>=0 else 0):(ind+r+1)])
        
        if i > upper_bound:
            LB_sum = LB_sum+ (i-upper_bound)**2
        elif i<lower_bound:
            LB_sum = LB_sum + (i-lower_bound)**2
    
    return np.sqrt(LB_sum)
```

### Example

```python
s1 = df.iloc[152,:-5].values.astype(float)
s2 = df.iloc[683,:-5].values.astype(float)
size = s1.shape[0]
r = 5
```

```python
LB_Keogh(s1, s2, size, r)
```

```python
LB_Keogh_py(s1, s2, r)
```

## Performance comparison

#### python

```python
%%timeit
LB_Keogh(s1, s2, size, r)
```

## Building a module

### output a .pyx file

```python
%%writefile dtw.pyx
cimport cython
import numpy as cnp
cimport numpy as cnp
from libc.math cimport sqrt, pow
cnp.import_array()

boundscheck=False
wraparound=False
cdef double minimum(double[:]  s, int i, int size, int r):

    cdef int lower = max(0, i - r)
    cdef int upper = min(size, i + r + 1)
    cdef int j
    cdef double m = s[lower]
    for j in range(lower, upper):
        if s[j] < m:
            m = s[j]
    return m

boundscheck=False
wraparound=False
cdef double maximum(double[:]  s, int i, int size, int r):

    cdef int lower = max(0, i - r)
    cdef int upper = min(size, i + r + 1)
    cdef int j
    cdef double m = s[lower]
    for j in range(lower, upper):
        if s[j] > m:
            m = s[j]
    return m  

boundscheck=False
wraparound=False

boundscheck=False
wraparound=False
def LB_Keogh(double[:] s1, double[:] s2, int size, int r):
    cdef double LB_sum = 0
    cdef double curr, lower, upper
    cdef int i
    for i in range(size):
        
        curr = s1[i]
        lower = minimum(s2, i, size, r)
        upper = maximum(s2, i, size, r)

        if curr > upper:
            LB_sum += pow((curr-upper), 2.0)
        elif curr < lower:
            LB_sum += pow((curr-lower), 2.0)


    return sqrt(LB_sum)
```

### output a setup.py file
Will be used to build a cython module

```python
%%writefile setup.py
import numpy as np
from distutils.core import setup
from Cython.Build import cythonize
from distutils.extension import Extension

extensions = [
    Extension("*", ["*.pyx"],
        include_dirs=[np.get_include()],
        libraries=[])
]

setup(
    name="DTW",
    ext_modules=cythonize(extensions),
)
```
